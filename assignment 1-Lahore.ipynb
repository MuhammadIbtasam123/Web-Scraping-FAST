```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
# scrapper for lahore campus
URL_Lahore="http://lhr.nu.edu.pk/faculty/"
r_Lahore = requests.get(URL_Lahore)
print (r_Lahore)
```

```python
soupLahore = BeautifulSoup(r_Lahore.content, 'html.parser')
print(soupLahore.prettify())
```

```python
f = soupLahore.findAll('div',class_="facultyCard")
print(f)
# print(len(f))
Dept = soupLahore.find('div',class_="container")
print(Dept)
```

```python
Id = []
Name = []
Designation = []
HEC_Approved = []
Highest_Education = []
Email = []
Department = []
Extension = []
ImageUrl = []

# 1- getting the Id (integer)
for i in range (len(f)):
    faculty = f[i]
    try:
        Id.append(int(faculty.find('a',class_='faculty-link')['href'].split('/')[-1]))
    except:
        Id.append(-1)


# 2- Getting the Name (string)
for i in range (len(f)):
    faculty = f[i]
    Name.append(str(faculty.find('h5',class_='text-center').text))


# 3- Getting the Designation (string)
for i in range (len(f)):
    faculty = f[i]
    Designation.append(str(faculty.find('p',class_='small text-center font-italic').text.strip().split('\n')[0]))
    
# 6- Getting the email (EMAIL)
for i in range (len(f)):
    faculty = f[i]    
    Email.append(faculty.find('p',class_='mb-0 text-center').text)
    
# 9- getting the ImageURL (URL)
for i in range (len(f)):
    faculty = f[i]
    ImageUrl.append('https://lhr.nu.edu.pk' + faculty.find('img',class_='card-img-top')['src'])
    
```

```python
# 7- getting the Department (string)

ArrayOfDepartments = []

# getting the no of departments and store in the list

lengthOf = len(Dept.find('div',class_='text-center white').findAll('a'))
for i in range(lengthOf):
    ArrayOfDepartments.append(Dept.find('div',class_='text-center white').findAll('a')[i].text)
print(ArrayOfDepartments)

# now for every department we will find how many instructors are there and store the length
DeptOf_FSC = soupLahore.find('div',id="fsc")
DeptOf_EE = soupLahore.find('div',id="ee")
DeptOf_CV = soupLahore.find('div',id="cv")
DeptOf_FSM = soupLahore.find('div',id="fsm")
DeptOf_SS = soupLahore.find('div',id="ss")

# print(DeptOf_FSC)
# print(DeptOf_EE)
# print(DeptOf_CV)
# print(DeptOf_FSM)
# print(DeptOf_SS)

lengthOf_FSC = len(DeptOf_FSC.findAll('h5',class_='text-center'))
lengthOf_EE = len(DeptOf_EE.findAll('h5',class_='text-center'))
lengthOf_CV = len(DeptOf_CV.findAll('h5',class_='text-center'))
lengthOf_FSM = len(DeptOf_FSM.findAll('h5',class_='text-center'))
lengthOf_SS = len(DeptOf_SS.findAll('h5',class_='text-center'))

# print(lengthOf_FSC)
# print(lengthOf_EE)
# print(lengthOf_CV)
# print(lengthOf_FSM)
# print(lengthOf_SS)

# now i have to make a loop that iterate till length of faculty and append the deoartments

for i in range(lengthOf_FSC):
    Department.append(ArrayOfDepartments[0])

for i in range(lengthOf_EE):
    Department.append(ArrayOfDepartments[1])
    
for i in range(lengthOf_CV):
    Department.append(ArrayOfDepartments[2])
    
for i in range(lengthOf_FSM):
    Department.append(ArrayOfDepartments[3])

for i in range(lengthOf_SS):
    Department.append(ArrayOfDepartments[4])
    
# for i in range(lengthOf_FSC):
#     print(DeptOfFSC.findAll('h5',class_='text-center')[i].text)
# print(lengthOf_FSC)



# print(Department.clear())
print(len(Department))
print(Department)

```

```python
# 8- getting the Extension (integer)

for i in range(len(Id)):
    Next_EXT_URL = 'https://lhr.nu.edu.pk/fsc/facultyProfile/'+str(Id[i])
    r_Next_EXT_URL_Lahore = requests.get(Next_EXT_URL)
    soup_Next_EXT_URL_Lahore = BeautifulSoup(r_Next_EXT_URL_Lahore.content, 'html.parser')
    # print(soup_Next_EXT_URL_Lahore.prettify())
    try:
        FacultyExt = soup_Next_EXT_URL_Lahore.find('div',class_='card-body')
        Extension.append(int(FacultyExt.find('p',class_='mt-0 text-center').text.split('Ext:')[1]))
    except:
        Extension.append('none')

print(len(Extension))
# Extension.clear()
# Extension = Extension[:184]
print(Extension)
```
