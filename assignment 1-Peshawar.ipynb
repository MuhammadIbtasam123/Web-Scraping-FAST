```python
#************************* PESHAWAR CAMPUS*************************************
import requests 
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
# scrapper for lahore campus
URL_PeshawarCS="http://pwr.nu.edu.pk/cs-faculty/"
URL_PeshawarEE="http://pwr.nu.edu.pk/ee-faculty/"
URL_PeshawarSH="http://pwr.nu.edu.pk/sh-faculty/"
r_PeshawarCS = requests.get(URL_PeshawarCS)
r_PeshawarEE = requests.get(URL_PeshawarEE)
r_PeshawarSH = requests.get(URL_PeshawarSH)
print (r_PeshawarCS)
print (r_PeshawarEE)
print (r_PeshawarSH)
```

```python
#    5- Getting the Highest Education (string)

length_of_CS = len(soupPeshawarCS.findAll('h4',class_="text-uppercase font-raleway font-weight-600 m-0"))
length_of_EE = len(soupPeshawarEE.findAll('h4',class_="text-uppercase font-raleway font-weight-600 m-0"))
length_of_SH = len(soupPeshawarSH.findAll('h4',class_="text-uppercase font-raleway font-weight-600 m-0"))
total = length_of_CS + length_of_EE + length_of_SH
print(length_of_CS)
print(length_of_EE)
print(length_of_SH)

def Calculate_Education(dept,ID,start,end):
    for i in range(start,end):
        Next_EXT_URL_CS = 'http://pwr.nu.edu.pk/'+dept+'-faculty-profile/?'+dept+'-profile='+str(ID[i])
        r_Next_EXT_URL_CS = requests.get(Next_EXT_URL_CS)
        soup_Next_EXT_URL_CS = BeautifulSoup(r_Next_EXT_URL_CS.content, 'html.parser')
        education = soup_Next_EXT_URL_CS.findAll('div',{'id':'education'})
        Highest_Education.append(str(education[0].li.text))
    

Calculate_Education('cs',Id,0,length_of_CS)
Calculate_Education('ee',Id,length_of_CS,length_of_CS+length_of_EE)
Calculate_Education('sh',Id,length_of_CS+length_of_EE,total)
```

```python
# 8- getting the Extension (integer)

length_of_CS = len(soupPeshawarCS.findAll('h4',class_="text-uppercase font-raleway font-weight-600 m-0"))
length_of_EE = len(soupPeshawarEE.findAll('h4',class_="text-uppercase font-raleway font-weight-600 m-0"))
length_of_SH = len(soupPeshawarSH.findAll('h4',class_="text-uppercase font-raleway font-weight-600 m-0"))
total = length_of_CS + length_of_EE + length_of_SH
print(length_of_CS)
print(length_of_EE)
print(length_of_SH)


def Calculate_Extension(dept,ID,start,end):
    for i in range(start,end):
        Next_EXT_URL_CS = 'http://pwr.nu.edu.pk/'+dept+'-faculty-profile/?'+dept+'-profile='+str(ID[i])
        r_Next_EXT_URL_CS = requests.get(Next_EXT_URL_CS)
        soup_Next_EXT_URL_CS = BeautifulSoup(r_Next_EXT_URL_CS.content, 'html.parser')
        FacultyExt = soup_Next_EXT_URL_CS.find('li',class_='m-0 pl-10 pr-10')
        Extension.append(int(FacultyExt.a.text.split("Ext ")[-1]))
    

Calculate_Extension('cs',Id,0,length_of_CS)
Calculate_Extension('ee',Id,length_of_CS,length_of_CS+length_of_EE)
Calculate_Extension('sh',Id,length_of_CS+length_of_EE,total)
    
```
